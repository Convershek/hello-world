10)	Ingresar números naturales hasta que el número ingresado sea 0. Mostrar cual fue el menor número ingresado (no cuenta el 0) y mostrar cuantos fueron pares y cuantos impares.

```
#Magic starts here
pares = 0
impares = 0
n = int(input("Ingrese numero: "))
menor = n
while n != 0:
  if (n % 2 == 0):
    pares = pares + 1;
  else:
    impares = impares + 1;  
  if (n < menor):
    menor = n;
  n = int(input("Ingrese numero: "))

print(menor)
print(pares)
print(impares)
  
```
